```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#한글 폰트해결
from matplotlib import font_manager, rc
font_path = "./malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

file_path = './game.csv'
df = pd.read_csv(file_path)
GL=df.loc[:,['Genre','Global_Sales']] #글로벌 판매량 기준
GLYear = df.loc[:,['Year','Genre','Global_Sales']] #연도포함
print('자료수:', len(GL))
print(df.head())

print(GLYear.head())
print('\n')
```

```python
#groupby를 이용한 데이터 분류/총판매량, 평균, 표준편차, 최대 판매량 표시
groupedGL = GL.groupby(['Genre'])
print(groupedGL)

sumGL = groupedGL.sum()
print('총 판매량')
print(sumGL)
print('\n')

averageGL = groupedGL.mean()
print('평균')
print(averageGL)
print('\n')

stdGL = groupedGL.std()
print('표준편차')
print(stdGL)
print('\n')

MaxGL = groupedGL.max()
print('최대 판매량')
print(MaxGL)
print('\n')
```

```python
grouped_2 = GLYear.groupby(['Year','Genre'])
print('연도별 장르 글로벌 매출액')
for key, group in grouped_2:
    print('* key :',key)
    print('* number :', len(group))
    print(group.head())
    print('총 판매량:',group['Global_Sales'].sum())
    print('평균:',group['Global_Sales'].mean())
    print('표준 편차:',group['Global_Sales'].std())
    print('최고 판매량:',group['Global_Sales'].max())
    print('\n')
```

```python
pivot_df = df.pivot_table(index='Genre', columns='Year', values='Global_Sales', fill_value=0, aggfunc='sum')
#장르를 인덱스로 하여 연도별 해당 장르의 판매량의 총합, 빈 값은 0으로 채우고 value값은 판매량의 총합으로 한다.
print('글로벌 판매량 정리')
pivot=pivot_df
pivot_df=pivot_df.T
print(pivot_df)
```

```python
#각 지역별 판매량 정리
pivot_dfNA = df.pivot_table(index='Genre', columns='Year', values='NA_Sales', fill_value=0, aggfunc='sum')
print('NA 판매량 정리')
pivot_dfNA=pivot_dfNA.T
print(pivot_df)

pivot_dfEU = df.pivot_table(index='Genre', columns='Year', values='EU_Sales', fill_value=0, aggfunc='sum')
print('EU 판매량 정리')
pivot_dfEU=pivot_dfEU.T
print(pivot_df)

pivot_dfJP = df.pivot_table(index='Genre', columns='Year', values='JP_Sales', fill_value=0, aggfunc='sum')
print('JP 판매량 정리')
pivot_dfJP=pivot_dfJP.T
print(pivot_df)

pivot_dfOT = df.pivot_table(index='Genre', columns='Year', values='Other_Sales', fill_value=0, aggfunc='sum')
print('OT 판매량 정리')
pivot_dfOT=pivot_dfOT.T
print(pivot_df)
```

```python
#각 자료 그래프
ax1=pivot_df.plot(figsize=(16,5))
ax1.set_title('연도별 각 장르의 총 판매량', size =20)

axNA=pivot_dfNA.plot(figsize=(16,5))
axNA.set_title('연도별 NA지역 각 장르의 총 판매량', size =20)

axEU=pivot_dfEU.plot(figsize=(16,5))
axEU.set_title('연도별 EU지역 각 장르의 총 판매량', size =20)

axJP=pivot_dfJP.plot(figsize=(16,5))
axJP.set_title('연도별 JP지역 각 장르의 총 판매량', size =20)

axOT=pivot_dfOT.plot(figsize=(16,5))
axOT.set_title('연도별 Other지역 각 장르의 총 판매량', size =20)

plt.style.use('ggplot')
ax2=sumGL.plot(figsize=(16,5),kind='bar')
ax2.set_title('장르별 총 판매량', size = 20)

ax3 = MaxGL.plot(figsize=(16,5), kind='bar', color='green')
ax3.set_title('장르별 최고 판매량', size = 20)
```
